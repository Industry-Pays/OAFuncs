```python
import OAFuncs as oaf
functions = [f for f in dir(oaf) if callable(getattr(oaf, f))]
print(functions)
# 获取每个函数的调用方法，因为比如可能是：oaf.oa_file.make_folder
for f in functions:
    print(f)
    print(getattr(oaf, f).__doc__)
    # 获取调用路径，中间可能有多个点
    print(getattr(oaf, f).__module__+'.'+f)
    # 输出函数的每个传入参数，不需要内部变量
    params = getattr(oaf, f).__code__.co_varnames
    params = [p for p in params if p != 'args' and p != 'kwargs']
    print(params)
```

```python
def query(libname='OAFuncs'):
    # 将libname转化为库对象，再传入dir函数
    funcs = [f for f in dir(f'{libname}') if callable(
        getattr(f'{libname}', f))]
    print('函数数量：')
    print(len(funcs))
    print('函数列表：')
    print(funcs)

oaf.oa_help.query()
oaf.oa_help.use('get_var')
```
